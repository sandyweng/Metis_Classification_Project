Prediction of Kickstarter Campaign Success

My goal is to create a model to predict whether the Kickstarter campaign will be successful based on features like the goal of the project, the duration of the project, and the day/month of when the project launched. I started with a Logistic Regression model as my baseline model. This model produced a F-beta score of 0.703. The confusion matrix depicts the positive class(successful project) vs negative class(failed project). This figure suggests that the model is getting a lot of True Positives but still needs work on False Positives(this one will cost the company money because they are thinking the project will succeed when it actually failed).
